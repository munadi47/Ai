```python
import pandas as pd
import numpy as np
import matplotlib.pyplot
```

```python
data = pd.read_csv('train.csv')
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.dtypes
```

```python
# Handling missing value (mode = modus mengambil nilai terbanyak)
data['Gender'] = data['Gender'].fillna(data['Gender'].mode()[0])

# Handling missing value (mean = modus mengambil nilai rata rata)
data['LoanAmount'] = data['LoanAmount'].fillna(data['LoanAmount'].mean())
```

```python
data.isnull().sum()
```

```python
#object pake mode, klo numeric/float pake mean
data['Married'] = data['Married'].fillna(data['Married'].mode()[0])
data['Dependents'] = data['Dependents'].fillna(data['Dependents'].mode()[0])
data['Self_Employed'] = data['Self_Employed'].fillna(data['Self_Employed'].mode()[0])
data['Loan_Amount_Term'] = data['Loan_Amount_Term'].fillna(data['Loan_Amount_Term'].mean())
data['Credit_History'] = data['Credit_History'].fillna(data['Credit_History'].mean())
```

```python
data.isnull().sum()
```

```python
print(pd.value_counts(data.Gender))
print(pd.value_counts(data.Education))
print(pd.value_counts(data.Married))
print(pd.value_counts(data.Dependents))
print(pd.value_counts(data.Self_Employed))
print(pd.value_counts(data.Property_Area))

#setelah ini akan dilakukan transformasi yang mengubah object
```

```python
map_gender = dict(zip(set(data['Gender']), [1,2]))
data['Gender'] = data['Gender'].apply(lambda x:map_gender[x])
print(map_gender)
```

```python
map_edu = dict(zip(set(data['Education']), [1,2]))
data['Education'] = data['Education'].apply(lambda x:map_edu[x])
print(map_edu)
```

```python
map_edu = dict(zip(set(data['Dependents']), [1,2,3,4]))
data['Dependents'] = data['Dependents'].apply(lambda x:map_edu[x])
print(map_edu)
```

```python
map_edu = dict(zip(set(data['Self_Employed']), [1,2]))
data['Self_Employed'] = data['Self_Employed'].apply(lambda x:map_edu[x])
print(map_edu)
```

```python
map_edu = dict(zip(set(data['Property_Area']), [1,2,3]))
data['Property_Area'] = data['Property_Area'].apply(lambda x:map_edu[x])
print(map_edu)
```
